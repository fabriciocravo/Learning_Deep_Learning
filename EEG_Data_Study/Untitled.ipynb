```python
# Necessary imports

import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
import numpy as np
```

```python
d1 = pd.read_csv('archive/s00.csv')
print(d1)
```

The collums in the dataset are not well-named, but they represent the different electrode channels. 
Maybe it's worth cleaning it up?

```python
# Plot just one of the 
for feature in d1:
    plt.plot(d1[feature])
    first_feature = feature
    plt.show()
    break
```

```python
# Fourrier transform given a time windown

t = 0
T = 5000
while(t < 30000):
    data = d1[first_feature].iloc[t:t+T+1].to_numpy()
    fty = fft(data)
    N = len(fty)
    plt.plot(np.abs(fty[0:N//2]))
    plt.show()
    t = t + T
    
```
